```python
month=[]
def month_con(dates):
    for date in dates:
        if re.search('^12',str(date)):
            month.append('Dec')
        elif re.search('^11',str(date)):
            month.append('Nov')
        elif re.search('^10',str(date)):
            month.append('Oct')
        elif re.search('^9',str(date)):
            month.append('Sept')
        elif re.search('^1',str(date)):
            month.append('Jan')
        else:
            month.append('null')
```

```python
month_con(df['Date'])
```

```python
month
```

```python
df1=df.assign(Month=month)
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
df1['Month'].unique()
```

```python
year=[]
def year_con(dates):
    for date in dates:
        if re.search('2021$',str(date)):
            year.append('2021')
        elif re.search('2022$',str(date)):
            year.append('2022')
```

```python
year_con(df['Date'])
```

```python
year
```

```python
df1=df1.assign(Year=year)
```
